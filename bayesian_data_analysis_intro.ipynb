## Part I. Coin Flipping

We begin by calculating a very simple system: coin tossing with a possibly biased coin. Let $H$ be the parameter which quantifies the probability of “heads”. For an unbiased coin, we should have $H = 0.5$. A coin with heads on both sides of the coin will have $H = 1$. For coin flipping, the probability of a set of results (data) is given by a binomial distribution. 

Here, we develop python code that can simulate coin tossing and calculate posterior distributions, i.e. $\text{prob}(H|D, I)$ for various assumed values of $H$. We then plot $\text{prob}(H|D, I)$ versus $H$ for a variety of number of trials $n$ and for several choices of the “true” value of $H$. 

We first investigate a uniform prior, i.e. $\text{prob}(X|I)$ = constant, before assuming a gaussian prior.

Using a uniform prior distribution, we can apply Bayes' Theorem directly to plot the posterior distribution at different values of $n$.

```python
NUM_ROWS = 6
NUM_COLS = 2

trials = np.array([0, 1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024])

def uniform_prior(input_values, *args):
    """
    Calculates uniform distribution between 0 and 1.
    
    Args:
        input_values: Array of input values
        
    Returns:
        Array of uniform outputs
    """
    return np.where((input_values >= 0) & (input_values <= 1), 1, 0)
    

def plot_heads_posterior_dist(H_true, trials, likelihood, prior=uniform_prior, *args):
    """
    Plots posterior distribution for coin flipping.
    
    Args:
        H_true: Assumed "true" probability of heads
        trials: Array of trial sizes
        prior: Prior distribution used
        *args: Additional arguments for prior
        
    Plots:
        Posterior distributions for each set of trials
    """
    total_heads = 0
    total_trials = 0
    fig, axes = plt.subplots(NUM_ROWS, NUM_COLS, figsize=(10, 20))
    H_values = np.linspace(0, 1, 100)
    posterior = prior(H_values, *args)
    
    # Run simulation
    for i in range(len(trials)):
        
        # Flip more coins
        while total_trials < trials[i]:
            total_heads += random.choices([0, 1], weights=(1 - H_true, H_true))[0]
            total_trials += 1
        
        # Calculate and normalize posterior
        posterior = likelihood(total_heads, total_trials, H_values) * prior(H_values, *args)
        normalization = integrate.simpson(posterior, H_values)
        posterior /= normalization
        
        # Plot posterior
        row = i // NUM_COLS
        col = i % NUM_COLS
        ax = axes[row, col]
        ax.plot(H_values, posterior)
        ax.grid()
        ax.set_xlabel('H')
        ax.set_ylabel('prob(H|D, I)')
        ax.set_title(f'Posterior Distribution for Heads (n={trials[i]})')
    plt.subplots_adjust(wspace=0.7, hspace=0.3)
    plt.tight_layout()
    plt.show()

H_true = 0.5
plot_heads_posterior_dist(H_true, trials, binom.pmf)
        
    
    
```

Using a Gaussian prior instead, we can show the effect of choosing a different true $H$ value.

$H = 0.5$

```python
def gaussian_prior(input_values, mean, stdev):
    """
    Calculates gaussian distribution given mean and standard deviation.
    
    Args:
        input_values: Array of input values
        mean: Mean of distribution
        stdev: Standard deviation of distribution
        
    Returns:
        Array of gaussian outputs
    """
    normal = norm(mean, stdev)
    return normal.pdf(input_values)
MEAN = 0.5
STDEV = 0.1
plot_heads_posterior_dist(H_true, trials, binom.pmf, gaussian_prior, MEAN, STDEV)
```

$H = 0.4$

```python
H_true = 0.4
plot_heads_posterior_dist(H_true, trials, binom.pmf, gaussian_prior, MEAN, STDEV)
```

```python
H_true = 0.8
plot_heads_posterior_dist(H_true, trials, binom.pmf, gaussian_prior, MEAN, STDEV)
```

## Part II. The Lighthouse Problem

A lighthouse is a distance $\alpha$ along the shore and a distance $\beta$ out to sea. It emits highly collimated flashes at random times and at random azimuths (a good lighthouse emits flashes at regular intervals, usually due to rotation of the lighthouse beam). A dense array of photodetectors is set up along the shore and each flash is recorded by a single photodetector. The data consist of the position $x_k$ of the illuminated photodetector for the kth flash. The distribution is shown in figure 2. For this part of the assignment, assume “true” values for $\alpha$ and $\beta$, say 1 km. Initially assume that $\beta$ is known and that the problem is to infer the posterior probability distribution of α from the data $\{x_k\}$.

Next we will consider the case where neither $\alpha$ nor $\beta$ is known beforehand and the challenge is to infer the most probable value from the data.

For the first part, we can use a uniform prior and a Cauchy distribution for the likelihood function to produce a posterior distribution for $\alpha$. For this situation, we assume that $\beta = 1$. We set the true position of the lighthouse to be at $(\alpha, \beta) = (1, 1)$ in order to generate the detection data randomly. The posterior distributions were plotted alongside the means to compare their values.

```python
# Figure parameters
NUM_ROWS = 5
NUM_COLS = 2

# Location of lighthouse
ALPHA_TRUE = 1.0
BETA_TRUE = 1.0

# Arbitrary bounds for alpha
ALPHA_MIN = -10.0
ALPHA_MAX = 10.0

STEP_SIZE = 0.05

trials = np.array([0, 1, 2, 4, 8, 16, 32, 64, 128, 256])

def alpha_uniform_prior(alpha_values):
    """
    Calculates uniform distribution between ALPHA_MIN and ALPHA_MAX.
    
    Args:
        input_values: Array of input values
        
    Returns:
        Array of uniform outputs
    """
    return np.where((alpha_values >= ALPHA_MIN) & (alpha_values <= ALPHA_MAX), 1, 0)

def alpha_likelihood(positions, alpha_values):
    """
    Calculates likelihood by using Cauchy distribution
    
    Args:
        positions: Array of positions of detections
        alpha_values: Range of alpha values to calculate likelihood for
        
    Returns:
        Array of likelihood values
    """
    likelihoods = np.empty(0)
    for alpha in alpha_values:
        pdfsum = np.sum(cauchy.logpdf(positions, loc=alpha))
        likelihoods = np.append(likelihoods, pdfsum)
    return np.exp(likelihoods)
    

def plot_lighthouse_posterior_dist(alpha_true, beta_true, trials):
    """
    Plots posterior distribution for lighthouse.
    
    Args:
        alpha_true: True alpha for lighthouse (used to generate data)
        beta_true: True beta for lighthouse (used to generate data)
        trials: Array of trial numbers
        
    Plots:
        Posterior distributions for each set of trials
    """
    alpha_values = np.arange(ALPHA_MIN, ALPHA_MAX, STEP_SIZE)
    angles = np.empty(0)
    total_trials = 0
    fig, axes = plt.subplots(NUM_ROWS, NUM_COLS, figsize=(10, 20))
    posterior = alpha_uniform_prior(alpha_values)
    
    # Run simulation
    for i in range(len(trials)):
        
        # Generate random positions
        while total_trials < trials[i]:
            angles = np.append(angles, random.uniform(-np.pi, np.pi))
            total_trials += 1
        positions = beta_true * np.tan(angles) + alpha_true
        
        # Calculate and normalize posterior
        likelihoods = alpha_likelihood(positions, alpha_values)
        posterior = likelihoods * alpha_uniform_prior(alpha_values)
        normalization = integrate.simpson(posterior, alpha_values)
        posterior /= normalization
        
        # Calculate average position
        average_position = np.mean(positions)
        print(f'Average position at {trials[i]} trials is {average_position}')
        
        # Plot posterior
        row = i // NUM_COLS
        col = i % NUM_COLS
        ax = axes[row, col]
        ax.plot(alpha_values, posterior, label='Posterior')
        ax.axvline(x=average_position, color='red', linestyle='--', label='Average Position')
        ax.grid()
        ax.set_xlabel('H')
        ax.set_ylabel('prob(H|D, I)')
        ax.set_title(f'Posterior Distribution for Alpha (n={trials[i]})')
        ax.legend()
    plt.subplots_adjust(wspace=0.7, hspace=0.3)
    plt.tight_layout()
    plt.show()

plot_lighthouse_posterior_dist(ALPHA_TRUE, BETA_TRUE, trials)
```

As seen in the graphs above, the average position is not necessarily the best measure of the alpha value since the mean is easily skewed by any far out positions which are much further from all other values, which are typically much closer to alpha. 

For the second part, we plot the contour of the posterior distributions for the lighthouse when alpha and beta are both unknown.

```python
# Figure parameters
NUM_ROWS = 5
NUM_COLS = 2

# True position of lighthouse (used to generate data)
ALPHA_TRUE = -2.0
BETA_TRUE = 1.5

# Arbitrary bounds for alpha
ALPHA_MIN = -5.0
ALPHA_MAX = 5.0

# Arbitrary bounds for beta
BETA_MIN = STEP_SIZE = 0.05
BETA_MAX = 5.0

trials = np.array([0, 1, 2, 4, 8, 16, 32, 64, 128, 200])

def alpha_beta_uniform_prior(alpha_values, beta_values):
    """
    Calculates uniform distribution for alpha and beta.
    
    Args:
        alpha_values: Array of alpha values
        beta_values: Array of beta values
        
    Returns:
        Array of uniform outputs
    """
    return np.ones((np.size(beta_values), np.size(alpha_values)))

def alpha_beta_likelihood(positions, alpha_values, beta_values):
    """
    Calculates likelihood by using Cauchy distribution
    
    Args:
        positions: Array of positions of detections
        alpha_values: Range of alpha values to calculate likelihood for
        beta_values: Range of beta values to calculate likelihood for
        
    Returns:
        2-D Array of likelihood values
    """
    likelihoods = np.empty((0, len(alpha_values)), dtype=float)
    for beta in beta_values:
        alpha_likelihoods = np.empty(0)
        for alpha in alpha_values:
            pdfsum = np.sum(cauchy.logpdf(positions, loc=alpha, scale=beta))
            alpha_likelihoods = np.append(alpha_likelihoods, pdfsum)
        likelihoods = np.vstack((likelihoods, alpha_likelihoods))
    return np.exp(likelihoods)

def plot_lighthouse_posterior_cont(alpha_true, beta_true, trials):
    """
    Plots posterior distribution for lighthouse when alpha and beta are both unknown.
    
    Args:
        alpha_true: True alpha for lighthouse (used to generate data)
        beta_true: True beta for lighthouse (used to generate data)
        trials: Array of trial numbers
        
    Plots:
        Posterior distribution contours for each set of trials
    """
    alpha_values = np.arange(ALPHA_MIN, ALPHA_MAX, STEP_SIZE)
    beta_values = np.arange(BETA_MIN, BETA_MAX, STEP_SIZE)
    angles = np.empty(0)
    total_trials = 0
    fig, axes = plt.subplots(NUM_ROWS, NUM_COLS, figsize=(10, 20))
    posterior = alpha_uniform_prior(alpha_values)
    
    # Run simulation
    for i in range(len(trials)):
        
        # Generate random positions
        while total_trials < trials[i]:
            angles = np.append(angles, random.uniform(-np.pi, np.pi))
            total_trials += 1
        positions = beta_true * np.tan(angles) + alpha_true
        
        # Calculate and normalize 2-D posterior
        likelihoods = alpha_beta_likelihood(positions, alpha_values, beta_values)
        posterior = likelihoods * alpha_beta_uniform_prior(alpha_values, beta_values)
        if total_trials != 0:
            normalization = integrate.simpson(integrate.simpson(posterior, alpha_values), beta_values)
            posterior /= normalization
        
        alphas, betas = np.meshgrid(alpha_values, beta_values)

        # Plot posterior contour
        row = i // NUM_COLS
        col = i % NUM_COLS
        ax = axes[row, col]
        contour = ax.contourf(alphas, betas, posterior, levels=20)
        plt.colorbar(contour, ax=ax)
        ax.set_xlabel('Alpha')
        ax.set_ylabel('Beta')
        ax.grid()
        ax.set_title(f'Posterior Distribution for Alpha and Beta (n={trials[i]})')
    plt.subplots_adjust(wspace=0.7, hspace=0.3)
    plt.tight_layout()
    plt.show()

MEAN = ALPHA_TRUE
STDEV = BETA_TRUE
plot_lighthouse_posterior_cont(ALPHA_TRUE, BETA_TRUE, trials)
```
